```python
dados = pd.read_csv('C:\\Users\\laury\\Downloads\\igm_virgula.csv', encoding='latin-1', sep = ';')
dados.head()
```

```python
#verificando qtde de linhas e colunas
dados.shape
```

```python
# pedido para remover as duplicatas, no caso não há
dados = dados.drop_duplicates()
dados.shape
```

```python
#verificar qtde valores nulos existentes e retorna em sumário
dados.isnull().sum()
```

```python
#trabalhando os dados = criando colunas para marcar colunas que tinham valores nulos 
#dados[''] = np.where(dados[''].isnull(), 'null', 'n/a')

dados['participacao_transf_receita'] = np.where(dados['participacao_transf_receita'].isnull(), 'null', 'n/a')
dados['servidores'] = np.where(dados['servidores'].isnull(), 'null', 'n/a')
dados['comissionados_por_servidor'] = np.where(dados['comissionados_por_servidor'].isnull(), 'null', 'n/a')
dados['perc_pop_econ_ativa'] = np.where(dados['perc_pop_econ_ativa'].isnull(), 'null', 'n/a')
dados['gasto_pc_saude'] = np.where(dados['gasto_pc_saude'].isnull(), 'null', 'n/a')
dados['hab_p_medico'] = np.where(dados['hab_p_medico'].isnull(), 'null', 'n/a')
dados['exp_vida'] = np.where(dados['exp_vida'].isnull(), 'null', 'n/a')
dados['gasto_pc_educacao'] = np.where(dados['gasto_pc_educacao'].isnull(), 'null', 'n/a')
dados['exp_anos_estudo'] = np.where(dados['exp_anos_estudo'].isnull(), 'null', 'n/a')
dados['nota_ciencias'] = np.where(dados['nota_ciencias'].isnull(), 'null', 'n/a')
dados['nota_humanas'] = np.where(dados['nota_humanas'].isnull(), 'null', 'n/a')
dados['nota_linguagem'] = np.where(dados['nota_linguagem'].isnull(), 'null', 'n/a')
dados['nota_mat'] = np.where(dados['nota_mat'].isnull(), 'null', 'n/a')
dados['nota_redacao'] = np.where(dados['nota_redacao'].isnull(), 'null', 'n/a')
dados['idhm'] = np.where(dados['idhm'].isnull(), 'null', 'n/a')
dados['ranking_igm'] = np.where(dados['ranking_igm'].isnull(), 'null', 'n/a')
dados['indice_governanca'] = np.where(dados['indice_governanca'].isnull(), 'null', 'n/a')

# verificando a base
dados.head()

#Aqui temos o código para criar as colunas de validação se há ou não valores nulos. 
#Isso é importante para manter um histórico das alterações realizadas na base.
```
